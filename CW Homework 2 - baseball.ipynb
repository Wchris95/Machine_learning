```python
import pandas as pd

dataset = pd.read_csv("baseball-9192.csv")
dataset.head()
```

```python
from matplotlib import pyplot as plt
from IPython.display import display

feature_name = ["BattingAvg","OnBasePct", "Runs", "Hits", "Doubles", "Triples", "HomeRuns", "RBI", "Walks", "Strikeouts", "StolenBases"]
target_name = "Salary"

for i in feature_name:
    print("Feature Model: ", i)
    plt.scatter(dataset[i],dataset[target_name])
    plt.show()
```

```python
rbiFocus = dataset['RBI']
salary = dataset['Salary']

x = np.array(rbiFocus).reshape(-1,1)
y = np.array(salary)
```

```python
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error

print(model.intercept_)
print(model.coef_)
pred=model.predict(X_test)
predictions = pred.reshape(-1,1)
print('MSE:', mean_squared_error(y_test,predictions))
print('RMSE', np.sqrt(mean_squared_error(y_test,predictions)))
```

```python
number_df = pd.read_csv('data_mnist.csv')
number_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import seaborn as sns

y = number_df['label']
X = number_df.drop('label', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify = y)
```

```python
print("X train shape: ", X_train.shape[0:])
print("y train shape: ", y_train.shape[0:])
print("X test shape: ", X_test.shape[0:])
print("y test shape: ", y_test.shape[0:])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(fit_intercept=True,
                        multi_class = 'auto',
                         penalty = 'l1',
                         solver='saga',
                         max_iter = 1000,
                         C=50,
                         verbose=2,
                         n_jobs=5,
                         tol=0.001
                        )
clf
```

```python
clf.fit(X_train,y_train)
```

```python
clf.classes_
```

```python
clf.coef_.shape
```

```python
clf.coef_[0].round(2)
```

```python
clf.intercept_
```

```python
clf.n_iter_[0]
```

```python
score = clf.score(X_train, y_train)
score
```

```python
from sklearn.metrics import classification_report
y_pred = clf.predict(X_test)
print("Model with 0.001 tolerance:")
print(classification_report(y_test.values, y_pred))
```

```python
from sklearn import metrics

predictions = clf.predict(X_train)
cm= metrics.confusion_matrix(y_true=y_train, y_pred = predictions, labels = clf.classes_)
cm
```

```python
from matplotlib import pyplot as plt
from IPython.display import display

plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True,linewidths=.5, square = True, cmap = 'coolwarm');
plt.ylabel('Actual Label')
plt.xlabel('Predicted result')
plt.title('Accuracy Score: {0}'.format(score))
```

```python
test_score = clf.score(X_test, y_test)
test_score
```

```python
from sklearn import metrics

predictions = clf.predict(X_test)
cm= metrics.confusion_matrix(y_true=y_test, y_pred = predictions, labels = clf.classes_)
cm
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True,linewidths=.5, square = True, cmap = 'coolwarm');
plt.ylabel('Actual Label')
plt.xlabel('Predicted result')
plt.title('Accuracy Score: {0}'.format(test_score))
```

```python
size = 4
test_df = pd.read_csv('test_mnist.csv')
# Get sample and make predictions on them
images = test_df.sample(size).values
preds = clf.predict(images)
# Show images and their predictions. Need to reshape them to 28x28 arrays
def plot_random_digits(images, predictions=None):
    """Helper function to plot random images from the test data file."""
    if predictions is None:
        predictions = [None for _ in images]
    _, axes = plt.subplots(nrows=1, ncols=len(images), figsize=(10, 3))
    for ax, image, prediction in zip(axes, images, predictions):
        ax.set_axis_off()
        ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
        if prediction is not None:
            ax.set_title(f'Prediction: {prediction}')
plot_random_digits(images.reshape(-1, 28, 28), preds)
```
